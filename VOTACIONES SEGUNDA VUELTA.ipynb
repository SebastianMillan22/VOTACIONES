#INFORME DE VOTACIONES PRESIDENCIALES SEGUNDA VUELTA

####EN ESTE INFORME GRÁFICO SE REPRESENTA EL RESULTADO DADO EL DÍA 29 DE MAYO DEL 2022, AL MOMENTO DE TENER LA TOTALIDAD DE VOTOS DEL PAÍS FRENTE A CADA UNO DE SUS CANDIDATOS. 
####EN DONDE PODEMOS OBSERVAR: GUSTAVO PETRO FUE EL ELECTO PRESIDENTE CON 11 291 986 DE VOTOS. FRENTE A SU CONTRINCANTE ELECTORAL, RODOLFO HERNÁNDEZ, QUE ACUMULO UN TOTAL DE 10 604 337 VOTOS. 


```python
import matplotlib.pyplot as plt   # Libreria de Graficas

# GRAFICA DE BARRAS

eje_X = ["Gustavo Petro","Rodolfo Hernández"]
eje_Y = [11.2, 10.5]  #Millones de personas

# Creamos la grafica de barras
plt.bar(eje_X,eje_Y)

# Leyenda en eje Y
plt.ylabel("Cantidad de votos")

# Leyenda eje X
plt.xlabel("Candidatos Políticos")

# Titulo de la grafica
plt.title("Elecciones Presidenciales Segunda Vuelta")

# Mostrar Grafica
plt.show()
```
